# Nonlinear Methods of Manifold Learning
Linear methods like PCA do not perform well on data with non-linear relationships. These non-linear relationships are better represented on a manifold. Manifold learning performs dimensionality reduction by representing data as low-dimensional manifolds embedded in a higher-dimensional space. I will illustrate all methods on MNIST code.

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import load_digits
digits = load_digits()
```

## Multidimensional Scaling (MDS)
Performs PCA on a distance matrix instead of the covariance matrix, based on pairwise Euclidean distances between data points.

```python
def distance_matrix(X,k): #X is data and k are number of nearest neighbours
    def dist(a, b): #Euclidean distance formula
        return np.sqrt(sum((a - b)**2))

    X = digits.images[0]
    distances = np.array([[dist(p1, p2) for p2 in X] for p1 in X]) #calculate a distance matrix for every row in image with every other row

    #We do not need all the neighbours - choose k nearest 
    neighbors = np.zeros_like(distances) #create a matrix of same size as distances matrix
    sort_distances = np.argsort(distances, axis=1)[:,1:k+1] #returns indices that would sort an array in ascending order - and then choose the k closest neighbours
    
    for k,i in enumerate(sort_distances): # (0,0), (1,7), (2,6), (3,1), (4,4), (5,3)
        neighbors[k,i] = distances[k,i]
    #neighbours is now a distance matrix of the closest k nearest neighbours of X
    
    return neighbors
```

In MDS it is important to double center your distance matrix. Double centered matrix is a matrix with all entries in one row summing to zero, for all rows, and with all entries in one column summing to zero, for all columns. This is important for further operations(?) - eigen decomposition. 

```python
def center_dist_matrix(X): #center and return the distance metric
    n_samples = X.shape[0] #number of samples - rows

    meanrows = np.sum(X, axis=0) / n_samples #sum every column and divide by n_samples 
    meancols = (np.sum(X, axis=1)/n_samples)[:, np.newaxis] #sum every row and divide by n_samples and increase dimension by one - sorta like transpose
    meanall = meanrows.sum() / n_samples # Mean across all rows (entire matrix)

    X -= meanrows
    X -= meancols
    X += meanall

    return X
```

### MDS Algorithm
1. Compute the distance matrix for this dataset and center it
2. Apply pca on this distance matrix

```python
def mds(data, k): #k is the number of principle components
    dist_mat = center_dist_matrix(data) #set the neighbours to 6
    eig_val_cov, eig_vec_cov = np.linalg.eig(dist_mat)
    eig_pairs = [(np.abs(eig_val_cov[i]), eig_vec_cov[i]) for i in range(len(eig_val_cov))]
    eig_pairs = sorted(eig_pairs, key=(lambda x: x[0]), reverse=True)
    eig_pairs = np.array(eig_pairs)
    transform = np.hstack(
        [eig_pairs[i, 1].reshape(image.shape[1], 1) for i in range(k)]
    )
    new_img = np.dot(image, transform)
    return new_img, transform
```

```python
new_img, transform = mds(distance_matrix(digits.images[0],6),2)
plt.gray()
plt.matshow(digits.images[0])
plt.xlabel('original.png')

#new_img = np.real(new_img)
plt.matshow(new_img)
```

```python
def inverse_pca(img, components):
    reconstruct = np.dot(img, components.T).astype(int)
    return reconstruct #.reshape(-1,8,8)
    
for i in range(8):
    new_image, transform = mds(digits.images[2], i+1)
    reconstruct = inverse_pca(new_image, transform)
    plt.matshow(reconstruct)
    plt.xlabel('reconstruct'+str(i+1))

plt.matshow(digits.images[0])
plt.xlabel('original.png')
```

## Isomap
Isomap does MDS but uses the geodesic distance instead, which is the distance of the shortest path between data points.

```python
def isomap(X, k):
    # Compute shortest paths for geodesic distance
    from sklearn.utils.graph import graph_shortest_path
    graph = graph_shortest_path(X, directed=False)
    return mds(graph, k)

new_img, transform = isomap(digits.images[0],8)
plt.matshow(new_img)
```
